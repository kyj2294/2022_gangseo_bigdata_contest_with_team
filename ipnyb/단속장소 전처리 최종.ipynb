### 단속장소별 단속건수 추가

```python
df = df.drop('Unnamed: 0', axis = 1)
df.head()
```

```python
vc_df = df['단속장소'].value_counts()
```

```python
df['단속장소'].value_counts().index[1]
```

```python
df['단속장소'].value_counts()[1]
```

```python
df['단속건수'] = 0
```

```python
def set_how_much(df, df2):
    for i in range(len(df)):
        for j in range(len(df2)):
            if df['단속장소'][i].strip() == df2.index[j].strip():
                df['단속건수'][i] = df2[j]
    return df
```

```python
set_how_much(df, vc_df)
df
```

```python
df.to_csv('단속장소_단속시간대_단속건수추가.csv', encoding='cp949')
```

### x, y 값이 모두같은 단속장소 제거

단속장소 명이 다르게 기입되어 있지만 X, Y 좌표 모두 동일한 행들의
단속장소를 합한 후 중복 제거

```python
df_3['xy'] = 0
```

```python
for i in range(len(df_3)):
    df_3['xy'][i] = df_3['X'][i],',',df_3['Y'][i]
    df_3['xy'][i] = df_3['X'][i],df_3['Y'][i]
```

```python
for i in range(len(df_3) -1):
    if df_3['xy'][i] == df_3['xy'][i+1]:
        df_3['단속건수'][i] += df_3['단속건수'][i+1]

df_3.drop_duplicates('xy', keep = 'first', inplace = True)
```

```python
df_3.reset_index(inplace = True)
df_3.drop('index', axis = 1, inplace = True)
```

```python
df_3.drop('xy', axis = 1, inplace = True)
```
