```python
def cnt_how_many(df, df2):
    for i in range(len(df)):
        for j in range(len(df2)):
            if df['설치지점(도로명주소)'][i] == df2.index[j].strip():
                df['CCTV위치 개수'][i] = df2[j]
    return df
```
