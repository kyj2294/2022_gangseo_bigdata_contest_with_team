### x,y 좌표가 모두같은 컬럼제거

```python
df_0 = pd.read_csv('넘버링.csv', encoding = 'cp949')
```

```python
df_0['xy'] = 0
```

```python
for i in range(len(df_0)):
    df_0['xy'][i] = df_0['X'][i],',',df_0['Y'][i]
    df_0['xy'][i] = df_0['X'][i],df_0['Y'][i]
```

```python
df_0.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
df_0
```

```python
for i in range(len(df_0) -1):
    if df_0['xy'][i] == df_0['xy'][i+1]:
        df_0['단속건수'][i] += df_0['단속건수'][i+1]

df_0.drop_duplicates('xy', keep = 'first', inplace = True)
```

```python
df_0
```

```python
df_0.reset_index(inplace = True)
df_0.drop('index', axis = 1, inplace = True)
```

```python
for i in range(len(df_0)):
    df_0['number'][i] = df_0.index[i]
```

```python
df_0
```

```python
df_0.to_csv('단속장소_numbering.csv', encoding = 'cp949')
```

```python
def set_dis_to_next_num(df):
    for i in range(len(df)-1):
        d = hav_dis(df, i, i+1)
        df['dis_to_next_num'][i] = d
    return df
```
